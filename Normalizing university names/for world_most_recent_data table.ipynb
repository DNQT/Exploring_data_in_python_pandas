```python
world_most_recent_data = world_ranking.iloc[world_ranking.groupby('university_name')['year'].idxmax()].copy()
world_most_recent_data.head(3)
```

### Sezione 5: Normalizziamo  i nomi delle università della tabella shangai_most_recent con i nomi delle università di university_and_country.

```python
def normalize (column_value):
    
    for university in university_and_country['university_name']:
        if column_value == university:
            return column_value
        
    pattern = re.compile(column_value)
    for university in university_and_country['university_name']:
        occurrence = pattern.search(university)
        if occurrence:
            return university
        else:
            reverse_pattern = re.compile(university)
            reverse_occurrence =reverse_pattern.search(column_value)
            if reverse_occurrence:
                return university
            
```

```python
world_most_recent_data['university_name_normalized'] = world_most_recent_data['university_name'].apply(normalize)
world_most_recent_data.head(3)
```
