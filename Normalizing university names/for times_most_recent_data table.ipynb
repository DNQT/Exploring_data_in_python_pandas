### Sezione 5: Normalizziamo  i nomi delle università della tabella times_most_recent con i nomi delle università di university_and_country.

```python
def to_normalize (column_value):
    
    for university in university_and_country['university_name']:
        if column_value == university:
            return column_value
        
    pattern = re.compile(column_value)
    for university in university_and_country['university_name']:
        occurrence = pattern.search(university)
        if occurrence:
            return university
        else:
            reverse_pattern = re.compile(university)
            reverse_occurrence =reverse_pattern.search(column_value)
            if reverse_occurrence:
                return university
            
    
    
```

```python
times_most_recent_data['university_name_normalized'] =times_most_recent_data['university_name'].apply(to_normalize)
times_most_recent_data.head(3)
```
